# Task
Load the Iris dataset, inspect its shape, column names, first few rows, and summary statistics. Visualize the relationships between features using scatter plots, the distribution of values using histograms, and identify outliers using box plots.

## Load the dataset

### Subtask:
Load the Iris dataset into a pandas DataFrame.


```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a pandas DataFrame
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target variable
df_iris['species'] = iris.target

# Display the first few rows of the DataFrame
display(df_iris.head())
```

## Inspect the data

### Subtask:
Print the shape, column names, and first few rows of the `df_iris` DataFrame. Use `.info()` and `.describe()` to get summary statistics.


```python
print("Shape of the DataFrame:", df_iris.shape)
print("\nColumn names:", df_iris.columns)
print("\nFirst 5 rows:\n", df_iris.head())
print("\nDataFrame Info:")
df_iris.info()
print("\nDescriptive Statistics:")
display(df_iris.describe())
```

## Visualize relationships

### Subtask:
Create scatter plots to visualize the relationships between different features, differentiating by species.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot matrix colored by species
sns.pairplot(df_iris, hue='species')

# Add a title to the plot
plt.suptitle('Pairwise Relationships of Iris Features by Species', y=1.02)

# Display the plot
plt.show()
```

## Visualize distributions

### Subtask:
Use histograms to visualize the distribution of values for each feature.


```python
# Create a figure and a set of subplots
features = df_iris.columns[:-1]  # Exclude the 'species' column
n_features = len(features)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
axes = axes.flatten()

# Iterate through each feature and create a histogram
for i, feature in enumerate(features):
    axes[i].hist(df_iris[feature], bins=20)
    axes[i].set_title(feature)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

## Identify outliers

### Subtask:
Use box plots to identify outliers for each feature.


```python
# Create a figure and a set of subplots for the box plots
features = df_iris.columns[:-1]  # Exclude the 'species' column
n_features = len(features)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
axes = axes.flatten()

# Iterate through each feature and create a box plot
for i, feature in enumerate(features):
    axes[i].boxplot(df_iris[feature])
    axes[i].set_title(feature)
    axes[i].set_ylabel('Value') # Add a y-label for clarity

# Adjust layout to prevent titles and labels from overlapping
plt.tight_layout()

# Display the plots
plt.show()
```
